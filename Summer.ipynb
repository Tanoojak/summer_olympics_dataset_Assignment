# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
df=pd.read_csv("summer.csv")
```

```python
d=df['City'].unique()
for i in d:
    print(i)
len(d)
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
d=df[df['Medal']=='Gold']
data=[]
for sport in d['Sport'].unique():
    data.append([sport, len(d[d['Sport']==sport])])
pd.DataFrame(data,columns = ['Sport','Gold Medals']).sort_values(by='Gold Medals', ascending=False).head().plot(x = 'Sport', y = 'Gold Medals', kind = 'bar', figsize = (5,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]
for sport in df['Sport'].unique():
    data.append([sport, len(df[df['Sport']==sport])])
pd.DataFrame(data,columns = ['Sport','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Sport', y = 'Medals', kind = 'bar', figsize = (5,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data=[]
for athlete in df['Athlete'].unique():
    data.append([athlete, len(df[df['Athlete']==athlete])])    
pd.DataFrame(data,columns = ['Athlete','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Athlete', y = 'Medals', kind = 'bar', figsize = (5,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
d=df[df['Medal']=='Gold']
data=[]
for athlete in d['Athlete'].unique():
    data.append([athlete, len(d[d['Athlete']==athlete])])    
pd.DataFrame(data,columns = ['Athlete','Gold Medals']).sort_values(by='Gold Medals', ascending=False).head().plot(x = 'Athlete', y = 'Gold Medals', kind = 'bar', figsize = (5,5))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]
for event in df['Event'].unique():
    data.append([event, len(df[df['Event']==event])])
pd.DataFrame(data,columns = ['Event','no of players']).sort_values(by='no of players', ascending=False).head().plot(x = 'Event', y = 'no of players', kind = 'bar', figsize = (5,5))

```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
d=df[df['Medal']=='Gold']
d=d[d['Gender']!='Men']
data=[]
for sport in d['Sport'].unique():
    data.append([sport, len(d[d['Sport']==sport])])
pd.DataFrame(data,columns = ['Sport','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Sport', y = 'Medals', kind = 'bar', figsize = (5,5))

```
